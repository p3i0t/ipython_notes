# KL Divergence
## Kullback–Leibler divergence
In information theory, Kullback–Leibler divergence, KL divergence, relative entropy, is a measure of the difference or distance of two distributions $P(x)$ and $Q(x)$ over the same alphabet $ A_{X}$. The KL divergence between $P(x)$ and Q(x) is
$$D_{KL}(P||Q) = \sum_{x}P(x)\log\frac{P(x)}{Q(x)}$$

*KL divergence is not symetric, i.e: $D_{KL}(P||Q) \neq D_{KL}(Q||P)$.* 
*The relative entropy satisfies $D_{KL}(P||Q)\geq 0$ (Gibbs' inequality) with equality only $P = Q$.*

Intuitively, there are three cases:

* *If P is high and Q is high, we are happy*
* *If P is high and Q is low, we pay a price*
* *If P is low, we don't care*

## Convexity
KL divergence is convex. The following toy example give some intuition of its convexity.

```python
import matplotlib.pyplot as plt
import numpy as np

# KL divergence calculation: p, q are discrete tuples
def KL(p, q):
    entropy = 0.0
    for index, p_i in enumerate(p):
        entropy += p_i * np.log(p_i/q[index])
    return entropy

q_1 = [.5, .5]
q_2 = [.6, .4]
#for sample in range(0,1,)]
KLs_1 = [KL((sample+0.001, 0.999-sample), q_1) for sample in np.arange(0, 1, 0.01)]
KLs_2 = [KL((sample+0.001, 0.999-sample), q_2) for sample in np.arange(0, 1, 0.01)]

f, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(np.arange(0, 1, 0.01), KLs_1)
ax2.plot(np.arange(0, 1, 0.01), KLs_2)

plt.show()
```
